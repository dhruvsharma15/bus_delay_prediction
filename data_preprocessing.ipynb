# Pre-processing of the dataset

Dataset used - Bus delays and breakdowns(https://data.cityofnewyork.us/Transportation/Bus-Breakdown-and-Delays/ez4e-fazm)

```python
import pandas as pd

data_path = '../data/Bus_Breakdown_and_Delays.csv'
```

```python
dataset = pd.read_csv(data_path)
dataset
```

### Removing all the data points with School_Age_or_PreK = Pre-K, with How_Long_Delayed = NaN, and removing the irrelevant features

```python
dataset_1 = dataset.drop(dataset[dataset["School_Age_or_PreK"] == "Pre-K"].index)

dataset_2 = dataset_1.dropna(subset = ["How_Long_Delayed"])

irrelevant_features = ["School_Age_or_PreK", "Busbreakdown_ID", "Created_On", "Has_Contractor_Notified_Schools", "Has_Contractor_Notified_Parents", "Have_You_Alerted_OPT", "Informed_On", "Incident_Number", "Last_Updated_On"]
subsampled_data = dataset_2.drop(irrelevant_features, axis = 1)

print(dataset.shape, dataset_1.shape, dataset_2.shape, subsampled_data.shape) 
subsampled_data
```

### Reducing redundancy in the Bus_Company_Name by dicarding the redundant entries with different names

```python
for i, row in subsampled_data.iterrows():
    bus_comp = str(row["Bus_Company_Name"])
    bus_comp = bus_comp.split('.')[0]
    bus_comp = bus_comp.split(' ')[0]
    subsampled_data.at[i,"Bus_Company_Name"] = bus_comp
```

### Removing all those data points with Borough = NULL


As they constitute ~5% of the data, this means we can discard these data points without losing out much information.

```python
subsampled_data.dropna(subset = ["Boro"], inplace = True)
subsampled_data
```

### Seeing the distribution of the Breakdown_or_Running_late for its attribute values

```python
print((subsampled_data.Breakdown_or_Running_Late == 'Running Late').sum())
print((subsampled_data.Breakdown_or_Running_Late == 'Breakdown').sum())
```

#### Seeing the values, we can safely remove the Breakdown cases.

```python
subsampled_data = subsampled_data.drop(subsampled_data[subsampled_data["Breakdown_or_Running_Late"] == "Breakdown"].index)
subsampled_data.dropna(subset = ["Breakdown_or_Running_Late"], inplace = True)
subsampled_data
```

### Splitting Occured_On into two columns - Date and time

```python
date_time = subsampled_data["Occurred_On"].str.split(" ", n = 1, expand = True)
subsampled_data["date"] = date_time[0]
subsampled_data["time"] = date_time[1]
subsampled_data.drop(columns =["Occurred_On"], inplace = True)
subsampled_data
```

### Analysing the number of discrete values of the attributes for deciding to convert which all to categorical form

```python
print('school year ', len(subsampled_data['School_Year'].value_counts()))
print('run type ', len(subsampled_data['Run_Type'].value_counts()))
print('route number ', len(subsampled_data['Route_Number'].value_counts()))
print('reason ', len(subsampled_data['Reason'].value_counts()))
print('schools serviced ', len(subsampled_data['Schools_Serviced'].value_counts()))
print('borough ', len(subsampled_data['Boro'].value_counts()))
print('bus company name ', len(subsampled_data['Bus_Company_Name'].value_counts()))
```

### Converting School_Year, Run_Type, Reason, Boro, Breakdown_or_Running_Late to categorical format

```python
subsampled_data['School_Year'] = pd.Categorical(subsampled_data.School_Year)
subsampled_data['Run_Type'] = pd.Categorical(subsampled_data.Run_Type)
subsampled_data['Reason'] = pd.Categorical(subsampled_data.Reason)
subsampled_data['Boro'] = pd.Categorical(subsampled_data.Boro)
subsampled_data['Breakdown_or_Running_Late'] = pd.Categorical(subsampled_data.Breakdown_or_Running_Late)
subsampled_data
```

### Splitting data into training and testing datasets
All the school years of 2018-2019 are used as the testing data points, and rest as the training data points

```python
test = subsampled_data[subsampled_data["School_Year"] == "2018-2019"]
train = subsampled_data.drop(subsampled_data[subsampled_data["School_Year"] == "2018-2019"].index)

train.drop("School_Year", axis = 1, inplace=True)
test = test.drop("School_Year", axis = 1, inplace=True)
```

### Saving the preprocessed data to a csv file

```python
train.to_csv("../data/train.csv", sep='\t')
test.to_csv("../data/test.csv", sep='\t')
```
